## Data Analysis for Omega Search

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data7 = np.load("measurements_e7.npy", allow_pickle=True)
data3 = np.load("measurements_e3.npy", allow_pickle=True)
```

```python
data_item = data7.item()

omega_array = data_item['Omega_Array']
iter_array = data_item['Iteration_Array']

print(f'Observables : {data_item.keys()}')
print(f'Omega : {omega_array}')
print(f'Iterations : {iter_array}')
```

```python
plt.scatter(omega_array[:-1], iter_array[:-1])
plt.title('Iterations against Omega for Tol = 1e-7')
plt.xlabel('Omega')
plt.ylabel('Iterations')
plt.show()
```

```python
print(f'Optimal Omega = {omega_array[np.where(iter_array == np.min(iter_array))]}')
```

```python
data_item = data3.item()

omega_array = data_item['Omega_Array']
iter_array = data_item['Iteration_Array']

print(f'Observables : {data_item.keys()}')
print(f'Omega : {omega_array}')
print(f'Iterations : {iter_array}')
```

```python
plt.scatter(omega_array[:], iter_array[:])
plt.title('Iterations against Omega for Tol = 1e-3')
plt.xlabel('Omega')
plt.ylabel('Iterations')
plt.show()
```

```python
print(f'Optimal Omega = {omega_array[np.where(iter_array == np.min(iter_array))]}')
```
